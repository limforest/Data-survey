* hardy weinberg equilibrium <br /> <br />
![hardy_weinberg](https://cdn.kastatic.org/ka-perseus-images/010efe45f63d8da65c92d7af25c503c0bc91f105.png) <br /> <br />
A frequency : p <br />
C frequency : q <br />
p + q = 1 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; -> allele frequency <br />
(p+q)^2 = 1 <br />
p^2 + 2pq + q^2 = 1  &nbsp;&nbsp; -> genoytpe frequency

* mendelian inheritance(segregation) <br /> <br />
![mendelian segregation](https://cdn.kastatic.org/googleusercontent/jmNyR9CR0sa_aSbyGoUXP-YRtVhc7u04d-5z1hbRba7fjucZjfMhLNesU_GUkMDY3hEt6dgUKAsNyrh3hc3jxPU) <br /> <br />
p = 0.5 <br />
q = 0.5 <br />
p^2 = 0.25 <br />
2pq = 0.5 <br />
q^2 = 0.25 <br />
p^2:2pq:q^2 = 1:2:1

```
from itertools import combinations_with_replacement
```

```
combinations_with_replacement(dic_allele_frq.keys(), 2)
```

```
list(combinations_with_replacement(dic_allele_frq.keys(), 2)) #genotype의 가능한 모든 경우의 수를 만들 수 있음
```

```
list(combinations_with_replacement(dic_allele_frq.keys(), 3)) 

# 뒤에 쓰는 숫자 갯수만큼 가지는 모든 경우의 수들을 만들어주는 기능
# ('C', 'C', 'A')와 ('C', 'A', 'C')는 같은 걸로 봐줌
```

```
dic_gt_frq_est = {}
for p,q in list(combinations_with_replacement(dic_allele_frq.keys(), 2)) : # 만들어진 3종류 각각의 한 요소씩 p,q로 불러옴
  if p == q :
    gt_frq = dic_allele_frq[p]**2 # 요소 두개가 같은 건 homo라는 뜻이므로 제곱을 해줌
  else :
    gt_frq = dic_allele_frq[p] * dic_allele_frq[q] * 2 # else인 것은 요소 두개가 다른 C와 A인 경우인 hetero이므로 2pq를 해줌
  dic_gt_frq_est[p+q] = gt_frq 
```

```
dic_gt_frq_est
```

```
dic_test = {}
for p,q in list(combinations_with_replacement(dic_allele_frq.keys(), 2)) :
  dic_test[p+q] = 1
```

```
dic_test
```

```
for p,q in list(combinations_with_replacement(dic_allele_frq.keys(), 2)) :
  if p == q :
    gt_frq = dic_allele_frq[p]**2
  else :
    gt_frq = dic_allele_frq[p] * dic_allele_frq[q] * 2
  print(type(gt_frq), gt_frq)
```

```
gt_count_est = [x*23 for x in dic_gt_frq_est.values()] # 23은 genotype
```

```
gt_count_est
```

```
gt_count_obs = [list(df_vcf_food.loc[0].values).count(x) for x in dic_gt_frq_est.keys()]
```

```
gt_count_obs # it doens't have any hetero
```

* chi2 analysis <br /> <br />
<img src='https://i.ytimg.com/vi/VskmMgXmkMQ/hqdefault.jpg' />

```
chi_2 = np.sum(((np.array(gt_count_obs) - np.array(gt_count_est))**2) / gt_count_est)
```

```
chi_2
```

```
from scipy.stats import chi2
```

```
chi2.sf(chi_2,1)

# P value (-06은 소숫점을 앞으로 6번 이동하라는 뜻, 0.00000162...)
# sf = Segnificant Function
# our data do not support hardy weinberg equilibrium
```

```
pv = chi2.sf(chi_2, 1)
```
