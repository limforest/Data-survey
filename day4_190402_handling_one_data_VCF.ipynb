**vcf** <br />
 VCF란 reference file은 contig많이 지니는데, 이러한 ref file에 fastq file을 붙이고 read하여 call variation한 것임

< column >

* '#CHROM'  column : human일 때 chromosome number 사용(chr1, chr2 ..), 하지만 대부분의 경우 genome의 real number(=contig number, assembly한 단위)를 사용<br />
* 'POS(=position)' column : 그 position의 genome이 reference(ref)서열에서 alternative(alt)서열로 바뀐 것<br />
* 'QUAL(=quality)' : 이 값이 낮을 수록 depth는 높음, hetero는 보통 낮은 값을 지님
* 'INFO' : DP = 12라면, read가 ref에 12개 붙은 것으로, ref서열에 read는 하나만이 아닌 여러개가 붙을 수도 있음, DP의 값이 커질 수록 depth가 깊다고 표현
* 'FORMAT' : GT:PL = GenoType:Phared-scaLed
* 마지막 column(=SRR5470223.sorted.bam) : 0/1:38,0,255라고 표현되어 있다면, 이는 FORMAT과 같은 형식으로, 0/1은 hetero region임을 알려주는 genotype이다. 38,0,255는 phred-scaled(PL)로, REF:ALT = A:a일 때, AA,Aa,aa임을 뜻한다. 이때, 0에 가까운 값일 수록 양적으로 가장 알맞은 결과임을 알 수 있음, 여기서는 Aa 부분이 0이기 때문에 hetero가 가장 알맞은 결과임을 알 수 있음

< else >

* ref에 short read를 mapping해보았을 때, 같은쌍도 있지만 다른것도 있음 


< vcf에서 중요한것 >
* how many depth you have?
* MQ = mapping quality (INFO column에 있음)
* state calling : 0/1

```
def call_gt(x) :
  ref = x['REF']
  alt = x['ALT'].split(',') #alt는 가끔 2개이상 지닐 수도 있으니
  gts = np.array([ref]+alt)
  calls = x['SRR5470223.sorted.bam']
  call = calls.split(':')[0]
  ixs = np.array(call.split('/')).astype(int)
  gt = gts[ixs]
  return gt
```

```
df_vcf_snp_highdp.apply(call_gt, axis=1) #apply 버그발생
```

```
# 다른 방법 시도

gt_list = []
for ix in df_vcf_snp_highdp.index:
  x = df_vcf_snp_highdp.loc[ix]
  y = call_gt(x)
  gt_list.append(y)
```
