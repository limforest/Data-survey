# Download one SRR file from NCBI and make BAM file

```
# Downloading a SRR file
# SRR ID를 변수에 지정하면 VCF를 만드는 명령어들 속에 SRR ID를 직접 입력하지 않고 변수를 사용하면 되기 때문에 좀 더 편리하다
query_sra = 'SRR5470223'
sra_dir1 = query_sra[0:3]
sra_dir2 = query_sra[0:6]
sra_dir3 = query_sra 
sra_file = query_sra+'.sra'
print('Downloading',query_sra)
!wget ftp://ftp-trace.ncbi.nih.gov/sra/sra-instant/reads/ByRun/sra/{sra_dir1}/{sra_dir2}/{sra_dir3}/{sra_file}
```

```
# Converting SRR to fastq
! fastq-dump  --origfmt -I  --split-files --gzip {query_sra}.sra
# Mapping fastq to reference
! bwa mem ../references/GCA_004328685.1_ASM432868v1_genomic.fna {query_sra}_1.fastq.gz {query_sra}_2.fastq.gz -o {query_sra}.sam
# paired end mapping correction 
! samtools fixmate -O bam {query_sra}.sam {query_sra}.bam
# sort the alignment file
! samtools sort -o {query_sra}.sorted.bam  {query_sra}.bam
```

```
df_snp.head()

# #CHROM : chromosome id assign, 무슨 chromosome이 있는지 알아보려면 df_snp['#CHROM'].unique() 하면 됨
# QUAL, INFO : quality가 낮을 수록 depth(DP)의 수도 낮음
# Read depth 26개, VDB(variant distance bias)가 0.42면 약 10개는 다르게 나온 것으로, 이 서열들이 확실하지 않다는 것
## 16개는 G(REF)이면 10개는 T(ALT)이다. 이를 0/1로 표시, reference/variant, hetero 임을 알려줌
# 1/1 = variant/variant homo
```
