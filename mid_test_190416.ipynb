Data preparation
1. Download SRRID : SRR8892235 and convert the sra file into fastq (paired end)
2. Map the fastq to reference genome (ftp://ftp.ncbi.nlm.nih.gov/genomes/all/GCF/000/005/845/GCF_000005845.2_ASM584v2/GCF_000005845.2_ASM584v2_genomic.fna.gz)
3. Do paired-end mapping correction 
4. Do sort the alignment file
5. Call variations (SNP/INDEL) from BAM file <br />
*Hint : “bcftools call -mv -Ov -o” 

Missions 
1. Convert VCF file to excel file preserving headers (outfilename : SRR8892235.xlsx) 
2. Filter the VCF file with DP > 20 and convert it into excel (outfilename : SRR8892235.DP20.xlsx)
3. Extract genotypes from DP filtered VCF such as AA, TT, GT, TA and so on and save it into excel (outfilename : SRR8892235.DP20.GT.xlsx)

```
import pandas as pd

df_vcf = pd.read_csv('./SRR8892235.vcf.gz.mv.vcf',sep='\t',comment='#',header=None)
```

```
columns = !cat ./SRR8892235.vcf.gz.mv.vcf | grep '#' | tail -1
columns = columns[0].split('\t')
df_vcf.columns = columns
```

```
df_vcf.head()
```

```
df_vcf.to_excel('SRR8892235.xlsx')
```

```
m_indel = df_vcf['INFO'].apply(lambda x : x[0:5] == 'INDEL')
df_vcf[m_indel].head() 
```

```
df_vcf_snp = df_vcf[~m_indel] 
```

```
df_vcf_snp['INFO'].apply(lambda x : x.split(';')[0].replace('DP=','')).values.astype(int)
```

```
dp_dist = df_vcf_snp['INFO'].apply(lambda x : x.split(';')[0].replace('DP=','')).values.astype(int)
```

```
dp_dist = pd.Series(dp_dist)
```

```
m_dp = dp_dist.values > 20
```

```
df_vcf_snp_highdp = df_vcf_snp[m_dp]
```

```
df_vcf.shape, df_vcf_snp.shape, df_vcf_snp_highdp.shape
```

```
df_vcf_snp_highdp.to_excel('SRR8892235.DP20.xlsx')
```

```
df_vcf_snp_highdp.head()
```

```
df_vcf_snp_highdp.columns = columns
c_accessions = columns[9:]
```

```
import numpy as np
def call_gt(x):
  ref = x['REF']
  alt = x['ALT'].split(',')
  gts = np.array([ref]+alt)
  calls = x[c_accessions]
  gts_list = []
  for call in calls :
    gt = call.split(':')[0]
    ixs = np.array(gt.split('/')).astype(int) 
    gts_list.append(''.join(gts[ixs]))
  return gts_list
```

```
data = df_vcf_snp_highdp.apply(call_gt, axis=1) 
```

```
data = np.stack(data)
```

```
data.shape
```

```
df_gt_call = pd.DataFrame(data=data)
```

```
df_gt_call.columns = [x.split('.')[0] for x in c_accessions]
```

```
df_gt_call.head()
```

```
df_gt_call.to_excel('SRR8892235.DP20.GT.xlsx')
```
