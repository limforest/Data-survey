```
import numpy as np
def call_gt(x):
  ref = x['REF']
  alt = x['ALT'].split(',')
  gts = np.array([ref]+alt+['N'])
  calls = x[c_accessions]
  gts_list = []
  for call in calls :
    call = call.replace('.','-1') #-1 means gts's last one : gts의 맨 마지막에 있는 N으로 바꿔주라는 것
    gt = call.split(':')[0]
    ixs = np.array(gt.split('/')).astype(int)
    gts_list.append(''.join(gts[ixs]))
  return gts_list
```

```
data = df_vcf_food.apply(call_gt, axis=1)
```
